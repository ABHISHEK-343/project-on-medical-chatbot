```python
from tkinter import *
import os

class LoginApp:
    def __init__(self, root):
        self.root = root
        self.root.title("Account Login")
        self.root.geometry("300x250")
        self.username = StringVar()
        self.password = StringVar()
        self.username_verify = StringVar()
        self.password_verify = StringVar()

        self.main_screen()

    def main_screen(self):
        self.clear_screen()
        Label(self.root, text="Select Your Choice", bg="blue", width="300", height="2", font=("Calibri", 13)).pack()
        Label(self.root, text="").pack()
        Button(self.root, text="Login", height="2", width="30", command=self.login).pack()
        Label(self.root, text="").pack()
        Button(self.root, text="Register", height="2", width="30", command=self.register).pack()

    def clear_screen(self):
        for widget in self.root.pack_slaves():
            widget.destroy()

    def register(self):
        self.clear_screen()
        self.root.title("Register")
        Label(self.root, text="Please enter the details below", bg="blue").pack()
        Label(self.root, text="").pack()

        Label(self.root, text="Username * ").pack()
        Entry(self.root, textvariable=self.username).pack()

        Label(self.root, text="Password * ").pack()
        Entry(self.root, textvariable=self.password, show='*').pack()

        Label(self.root, text="").pack()
        Button(self.root, text="Register", width=10, height=1, bg="blue", command=self.register_user).pack()

    def register_user(self):
        username_info = self.username.get()
        password_info = self.password.get()

        if username_info and password_info:
            with open(username_info, "w") as file:
                file.write(username_info + "\n")
                file.write(password_info)
            self.username.set("")
            self.password.set("")

            Label(self.root, text="Registration Success", fg="green", font=("calibri", 11)).pack()
            Button(self.root, text="Click Here to Proceed", command=self.main_screen).pack()

    def login(self):
        self.clear_screen()
        self.root.title("Login")

        Label(self.root, text="Please enter details below to login").pack()
        Label(self.root, text="").pack()

        Label(self.root, text="Username * ").pack()
        Entry(self.root, textvariable=self.username_verify).pack()

        Label(self.root, text="Password * ").pack()
        Entry(self.root, textvariable=self.password_verify, show='*').pack()

        Label(self.root, text="").pack()
        Button(self.root, text="Login", width=10, height=1, command=self.login_verify).pack()

    def login_verify(self):
        username1 = self.username_verify.get()
        password1 = self.password_verify.get()

        if username1 in os.listdir():
            with open(username1, "r") as file:
                data = file.read().splitlines()
                if password1 == data[1]:
                    self.popup_message("Login Success")
                else:
                    self.popup_message("Invalid Password")
        else:
            self.popup_message("User Not Found")

    def popup_message(self, message):
        popup = Toplevel(self.root)
        popup.title("Message")
        popup.geometry("150x100")
        Label(popup, text=message).pack()
        Button(popup, text="OK", command=popup.destroy).pack()

if __name__ == "__main__":
    root = Tk()
    app = LoginApp(root)
    root.mainloop()
```
